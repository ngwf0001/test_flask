```python
import requests
from pprint import pprint
url = 'https://weimergeeks.com/examples/scraping/example1.html'
from bs4 import BeautifulSoup
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

heading = soup.h1
print(heading.text)
```

```python
for city_tag in soup.find_all('td', class_='city'):
    print(city_tag.text)
```

```python
citynames = [city_tag.text for city_tag in soup.find_all('td', class_='city')]
print(citynames)
```

```python
url = 'https://www.boxofficemojo.com/year/2024/'

soup = BeautifulSoup(requests.get(url).text, 'html.parser')
rows = soup.find('table').find_all('tr')[1:2]
for row in rows:
    cells = (row.find_all('td'))
    pprint(cells)
```

```python
[cell.text for cell in cells]
```
